```python
from transformers import AutoProcessor, MusicgenForConditionalGeneration
import torch
import numpy as np
from IPython.display import Audio

# GPU ì‚¬ìš© ì—¬ë¶€ í™•ì¸ ë° ì„¤ì •
device = "cuda" if torch.cuda.is_available() else "cpu"

# ìŒì•… ìƒì„± ëª¨ë¸ ì´ˆê¸°í™”
processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small").to(device)

def generate_music(summary, duration_seconds, base_tokens=256, base_duration=5):
    # Musicgen ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ ìŒì•… ìƒì„±
    # ì£¼ì–´ì§„ ì´ˆ ë‹¨ìœ„ ì‹œê°„ì„ ë°”íƒ•ìœ¼ë¡œ max_new_tokens ê°’ì„ ê³„ì‚°
    tokens_per_second = base_tokens / base_duration
    max_new_tokens = int(tokens_per_second * duration_seconds)

    inputs = processor(text=[summary], padding=True, return_tensors="pt").to(device)
    audio_values = model.generate(**inputs, max_new_tokens=max_new_tokens)
    sampling_rate = model.config.audio_encoder.sampling_rate
    
    # ì˜¤ë””ì˜¤ ë°ì´í„°ë¥¼ ìŠ¤íŠ¸ë¦¬ë°ìœ¼ë¡œ ë°˜í™˜
    audio_data = audio_values[0, 0].cpu().numpy()
    
    return audio_data, sampling_rate

# ì‚¬ìš©ìë¡œë¶€í„° ì›í•˜ëŠ” ìŒì•… ê¸¸ì´ë¥¼ ì´ˆ ë‹¨ìœ„ë¡œ ì…ë ¥ë°›ìŒ
duration_seconds = int(input("ìƒì„±í•  ìŒì•…ì˜ ê¸¸ì´ë¥¼ ì´ˆ ë‹¨ìœ„ë¡œ ì…ë ¥í•˜ì„¸ìš”: "))

# ìŒì•… ìƒì„±
audio_data, sampling_rate = generate_music(summary, duration_seconds, base_tokens=256, base_duration=5)


# ì˜¤ë””ì˜¤ ë°ì´í„° ì¬ìƒ (IPython Audio ê°ì²´ë¥¼ ì‚¬ìš©í•˜ì—¬ ì¬ìƒ)
Audio(data=audio_data, rate=sampling_rate)
```

```python
from transformers import pipeline, AutoProcessor, MusicgenForConditionalGeneration
import scipy.io.wavfile
import os

# ìŒì•… ìƒì„± ëª¨ë¸ ì´ˆê¸°í™”
processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")

def generate_music(summary, directory="./demo/"):
    # Musicgen ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ ìŒì•… ìƒì„±
    inputs = processor(text=[summary], padding=True, return_tensors="pt")
    audio_values = model.generate(**inputs, max_new_tokens=256)
    sampling_rate = model.config.audio_encoder.sampling_rate
    
    # íŒŒì¼ ì €ì¥
    existing_files = os.listdir(directory)
    files = [f for f in existing_files if f.startswith("music") and f.endswith(".wav")]
    n = len(files) + 1
    output_filename = f"{directory}music_{n}.wav"
    scipy.io.wavfile.write(output_filename, rate=sampling_rate, data=audio_values[0, 0].numpy())
    
    return output_filename


# ìŒì•… ìƒì„±
music_file = generate_music(summary)
print(f"Generated music is saved in: {music_file}")
```
